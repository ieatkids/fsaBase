### Incubator

* 基于市场数据的alpha，基本结构为二叉树，末端为variables(fields or functions)，其余结点为operators。
* 利用遗传规划可以进行alpha的搜索（进行中）

```python
f1 = incubator.AlphaTree('sub', 'AQ1', 'AQ2')
print(f1)
```

```python
f2 = incubator.AlphaTree('add', 'AQ1', 'BQ1')
print(f2)
```

```python
f3 = incubator.AlphaTree('div', f1, f2)
print(f3)
```

```python
arr = f3.getArray(marketDf)
print(arr)
```
